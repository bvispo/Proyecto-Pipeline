# 4. Data Preparation

```python
# Merge two DataFrames: cities and continents
cities_continents = pd.merge(df_cities, df_continents, left_on='iso2', right_on='Two_Letter_Country_Code')
cities_continents
```

```python
# Drop unnecessary columns
cities_continents.drop(columns=[
    'city','admin_name','capital','population', 'id',
    'Continent_Code', 'Country_Name', 'Two_Letter_Country_Code',
    'Three_Letter_Country_Code', 'Country_Number', "lat", "lng", "iso2", "iso3"],inplace=True)
```

```python
cities_continents
```

```python
cities_continents.isnull().sum()
```

```python
cities_continents_visualization = pd.merge(df_smart, cities_continents, left_on='City', right_on='city_A')
```

```python
cities_continents_visualization
```

```python
# Drop newunnecessary columns
cities_continents_visualization.drop(columns=['City','Country'],inplace=True)
cities_continents_visualization
```

```python
df = cities_continents_visualization.drop_duplicates()
```
